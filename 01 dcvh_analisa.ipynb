## Project credit card holder

### Objective:
- This project aims to create clusters defined according to the characteristics of credit card holders, grouping these holders and characterizing the clusters based on the occurrence of the variables used.

- Esse projeto tem como objetivo criar clusters definidos conforme as características dos titulares de cartão de crédito, agrupando esses titulares e caracterizando os clusters a partir da ocorrência das variáveis utilizadas.

### Data Origin:
- Dataset: https://www.kaggle.com/datasets/arjunbhasin2013/ccdata

- The sample dataset summarizes the usage behavior of about 9000 active credit card holders during the last 6 months. The file is at a customer level with 18 behavioral variables.

- O Dataset de amostra resume o comportamento de uso de cerca de 9000 titulares de cartão de crédito ativos durante os últimos 6 meses. O arquivo está em um nível de cliente com 18 variáveis ​​comportamentais.

- A seguir está o Dicionário de Dados para o conjunto de dados de Cartão de Crédito:-

    CUST_ID : Identificação do titular do cartão de crédito (categórico)

    BALANCE : Valor do saldo restante na conta para fazer compras

    BALANCE_FREQUENCY : Com que frequência o saldo é atualizado, pontuação entre 0 e 1 (1 = atualizado com frequência, 0 = não atualizado com frequência)

    PURCHASES : Quantidade de compras feitas na conta

    ONEOFF_PURCHASES : Valor máximo de compra feita de uma só vez

    INSTALLMENTS_PURCHASES : Valor da compra feita em parcelas

    CASH_ADVANCE : Dinheiro adiantado dado pelo usuário

    PURCHASES_FREQUENCY : Com que frequência as compras estão sendo feitas, pontuação entre 0 e 1 (1 = compradas com frequência, 0 = não compradas com frequência)

    ONEOFFPURCHASESFREQUENCY : Com que frequência as compras estão acontecendo de uma só vez (1 = compradas com frequência, 0 = não compradas com frequência)

    PURCHASESINSTALLMENTSFREQUENCY : Com que frequência as compras em parcelas estão sendo feitas (1 = feitas com frequência, 0 = não feito com frequência)

    CASHADVANCEFREQUENCY : Com que frequência o dinheiro adiantado é pago

    CASHADVANCETRX : Número de transações feitas com "Dinheiro adiantado"

    PURCHASES_TRX : Número de transações de compra feitas

    CREDIT_LIMIT : Limite do cartão de crédito para o usuário

    PAYMENTS : Valor do pagamento feito pelo usuário

    MINIMUM_PAYMENTS : Valor mínimo de pagamentos feitos pelo usuário

    PRCFULLPAYMENT : Porcentagem do pagamento integral pago pelo usuário

    TENURE : Prazo do serviço de cartão de crédito para o usuário

- Algumas variáveis possuem valores de grande dimensão, enquanto outras valores entre 0 e 1, vamos transformar essas variáveis para facilitar o trabalho mantendo uma dimensão mais próxima umas das outras.

```python
colunas_minmax = [
                'saldo_cred_restante',               
                'qtde_compras',
                'vr_max_compra_feita',
                'vr_compra_feita_parcelas',
                'dinh_adiantado_usu',
                'limite_cartao',
                'vr_pag_feito_usu',
                'vr_min_pagtos_feito_usu',
                ]

colunas_power = [
                'freq_atu_saldo',
                'freq_compras',
                'freq_compras_1vez',
                'freq_compras_parcelas',
                'freq_pag_dinh_adiantado',
                'trans_dinh_adiantado',
                'trans_compra_feita',
                '%pag_integral_feito_usu',
                'prazo_serv_cartao_usu'
                ]
```

```python
# definindo uma instância do columnTransformer para fazer uma verificação de transformação de variáveis
preprocessing = ColumnTransformer(
    [
        #("one-hot", OneHotEncoder(), colunas_onehot),
        #("standard", StandardScaler(), colunas_nro),
        ("minmax", MinMaxScaler(), colunas_minmax),
        ("power", PowerTransformer(), colunas_power),
    ]
)

preprocessing.fit_transform(df_card)
```

- Aqui os dados foram transformados, primeiro utilizando minmax para converter a dimensão das variáveis entre 0 e 1, melhorando a normalidade, diminuindo essa dimensão, e com power para melhorar a normalidade da distribuição de outras variáveis.

```python
# fazendo a transformação de variáveis com fit_transform e convertendo para um dataframe
df_card_trans = pd.DataFrame(
    preprocessing.fit_transform(df_card),
    columns=preprocessing.get_feature_names_out(), 
)
```

- Nessa fase as variáveis são transformadas através do fit_transform e convertidos em dataframe sendo armazenadas no dataframe df_card_trans para posterior junção com o dataframe de variáveis categóricas originais.

---

```python
# juntando o dataframe original dos dados categóricos com o dataframe gerado com variáveis transformadas
df_card_trans = pd.concat([df_card, df_card_trans], axis=1)
```

```python
# verificando a quantidade de linhas do dataframe original com o dataframe novo
print('df_card_trans', df_card_trans.shape)
print('df_card', df_card.shape)
```

```python
# verificando se existem nulos após a transformação
df_card_trans.isnull().sum()
```

```python
# apagando algumas linhas que possuem nulos
df_card_trans = df_card_trans.dropna()
```

```python
# verificando quantas linhas sobraram, o total continua satisfatório para este trabalho
df_card_trans.shape
```

```python
# verificando os dados das primeiras linhas dos dois dataframes juntados
df_card_trans.head(3)
```

```python
# plotando os histogramas das variáveis novamente para verificar se houve melhora na distribuição e também se houve
# alguma perda com a última limpeza das linhas com nulos
minmax_power_cols = df_card_trans.columns[
   df_card_trans.columns.str.startswith("minmax") | df_card_trans.columns.str.startswith("power")]

fig, axs = plt.subplots(nrows=3, ncols=6, figsize=(14, 14), tight_layout=True)

for ax, feature in zip(axs.flatten(), minmax_power_cols):
    sns.histplot(x=feature, ax=ax, data=df_card_trans, kde=True)

plt.show()
```
